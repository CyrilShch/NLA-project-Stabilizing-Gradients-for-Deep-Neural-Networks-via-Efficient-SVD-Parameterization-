```python
def load_data(model,dataset,layers,reflectors):
    name = 'data_'+model+'_dataset_'+dataset+'_L_'+str(layers)+'_rsize_'+str(reflectors)
    with open(name,'r') as file:
        lines = file.readlines()
        x = np.zeros(len(lines))
        y = np.zeros(len(lines))
        for i in range(len(lines)):
            data = np.array(lines[i].split(' '))
            x[i],_,y[i] = data.astype(float)
    return x,y
```

```python
model = 'LSTM'
dataset = 'add'
layers = 30
reflectors = 2
x_30,y_30 = load_data(model,dataset,layers,reflectors)
model = 'LSTM'
dataset = 'add'
layers = 100
reflectors = 2
x_100,y_100 = load_data(model,dataset,layers,reflectors)
model = 'LSTM'
dataset = 'add'
layers = 300
reflectors = 2
x_300,y_300 = load_data(model,dataset,layers,reflectors)
```

```python
plt.figure(figsize=(8, 8))
plt.plot(x_30[2:],y_30[2:],label='30')
plt.plot(x_100[2:],y_100[2:],label='100')
plt.plot(x_300[2:],y_300[2:],label='300')
plt.legend(fontsize=18)
plt.title(model+'_dataset_'+dataset,fontsize=18)
plt.ylabel("Calculation time", fontsize=18)
plt.xlabel("Length of array", fontsize=18)
#plt.yscale("log")
#plt.xscale("log")
plt.grid(True)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
```
